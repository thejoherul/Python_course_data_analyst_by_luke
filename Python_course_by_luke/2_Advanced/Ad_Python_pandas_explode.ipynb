```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df[['job_title_short', 'job_skills']].head(5)
```

```python
job_exploded = df.explode('job_skills')
job_exploded['job_skills'].value_counts().head(10).plot(kind='bar')
```

```python

skills_count = job_exploded.groupby(['job_title_short', 'job_skills']).size()
df_skills_count = skills_count.reset_index(name='skills_count')
df_skills_count =df_skills_count.sort_values(by='skills_count', ascending=False)

job_title = 'Data Scientist'
top_skill = 10
df_skills_final = df_skills_count[df_skills_count['job_title_short'] == job_title].head(top_skill)
df_skills_final.plot(kind='barh', x='job_skills', y='skills_count')
plt.xlabel('Job Posting Count')
plt.ylabel('')
plt.title(f'Top {top_skill} Skills for {job_title}s')
plt.gca().invert_yaxis()
plt.legend().set_visible(False)
plt.show()
```
