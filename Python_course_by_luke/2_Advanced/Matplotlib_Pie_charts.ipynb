```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_no_degree_mention'].value_counts().plot(kind='pie', startangle=90, autopct='%1.1f%%')
plt.title('Job_No_ Degree_Mention')
plt.ylabel("")
plt.show()
```

```python
fig, ax = plt.subplots(1, 3)

dict_column = {
    'job_health_insurance': 'Helth INS offer',
    'job_no_degree_mention': 'job req',
    'job_work_from_home': 'office or home'
}
for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle=90, autopct='%1.1f%%', labels=['False', 'True'])
    ax[i].set_title(title)
plt.show()
```
